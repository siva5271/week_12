```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler
```

```python
wine=load_wine()
wine.keys()
```

```python
df=pd.DataFrame(wine.data,columns=wine.feature_names)
df.info()
```

```python
scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(df)
df_scaled=pd.DataFrame(scaled_data,columns=df.columns)
df_scaled
```

```python
df_scaled['target']=wine.target
df_scaled
```

```python
X=df_scaled.drop(['target'],axis=1)
y=df_scaled['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
modelGaussian=GaussianNB()
modelGaussian.fit(X_train,y_train)
```

```python
modelGaussian.score(X_test,y_test)
```

```python
modelMultinomial=MultinomialNB()
modelMultinomial.fit(X_train,y_train)
```

```python
modelMultinomial.score(X_test,y_test)
```
