```python
import pandas as pd
```

```python
df=pd.read_csv('/home/siva/Downloads/py-master/ML/FeatureEngineering/3_outlier_IQR/Exercise/height_weight.csv')
df
```

### Handling weight column

```python
q1=df['weight'].quantile(0.25)
q3=df['weight'].quantile(0.75)
upper_limit=q3+(1.5*(q3-q1))
lower_limit=q1-(1.5*(q3-q1))
df2=df[(df['weight']>lower_limit)&(df['weight']<upper_limit)]
df2
```

### Taking care of height

```python
q1=df['height'].quantile(0.25)
q3=df['height'].quantile(0.75)
lower_limit=q1-(1.5*(q3-q1))
upper_limit=q3+(1.5*(q3-q1))
df3=df2[(df2['height']>lower_limit)&(df2['height']<upper_limit)]
df3
```
